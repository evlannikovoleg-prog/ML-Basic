## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Загрузка датасета
tips = sns.load_dataset('tips')

print("=" * 60)
print("📊 КОМПЛЕКСНЫЙ АНАЛИЗ ДАТАСЕТА TIPS")
print("=" * 60)

# 📊 ОБЩАЯ ХАРАКТЕРИСТИКА ДАННЫХ
print("\n1. 📋 ОБЩАЯ ХАРАКТЕРИСТИКА ДАННЫХ")
print("-" * 40)
print(f"Размер датасета: {tips.shape[0]} записей, {tips.shape[1]} признаков")
print(f"Период анализа: {', '.join(tips['day'].unique())}")
print("Пропуски в данных: отсутствуют")

# 💰 ОСНОВНЫЕ ФИНАНСОВЫЕ ПОКАЗАТЕЛИ
print("\n2. 💰 ОСНОВНЫЕ ФИНАНСОВЫЕ ПОКАЗАТЕЛИ")
print("-" * 40)

# Анализ total_bill
total_bill_stats = tips['total_bill'].describe()
print(f"\n📈 СЧЕТА (Total Bill):")
print(f"  • Средний счет: ${total_bill_stats['mean']:.2f}")
print(f"  • Медиана: ${total_bill_stats['50%']:.2f} (распределение скошено вправо)")
print(f"  • Диапазон: ${tips['total_bill'].min():.2f} - ${tips['total_bill'].max():.2f}")
print(f"  • Стандартное отклонение: ${total_bill_stats['std']:.2f} (высокая вариативность)")

# Анализ tip
tip_stats = tips['tip'].describe()
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100
print(f"\n💵 ЧАЕВЫЕ (Tip):")
print(f"  • Средние чаевые: ${tip_stats['mean']:.2f}")
print(f"  • Медиана: ${tip_stats['50%']:.2f}")
print(f"  • Диапазон: ${tips['tip'].min():.2f} - ${tips['tip'].max():.2f}")
print(f"  • Средний процент чаевых: {tips['tip_percentage'].mean():.1f}%")

# 📈 КЛЮЧЕВЫЕ ВЗАИМОСВЯЗИ
print("\n3. 📈 КЛЮЧЕВЫЕ ВЗАИМОСВЯЗИ")
print("-" * 40)

# Корреляция total_bill vs tip
correlation = tips['total_bill'].corr(tips['tip'])
slope, intercept, r_value, p_value, std_err = stats.linregress(tips['total_bill'], tips['tip'])
print(f"\n🔗 Total Bill vs Tip:")
print(f"  • Корреляция Пирсона: {correlation:.3f} (сильная положительная)")
print(f"  • R²: {r_value**2:.3f}")
print(f"  • p-value: {p_value:.4f} (статистически значимо)")

# Визуализация корреляции
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title(f'Total Bill vs Tip (корреляция: {correlation:.3f})')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

plt.subplot(1, 2, 2)
sns.scatterplot(data=tips, x='total_bill', y='tip_percentage', alpha=0.6)
plt.axhline(y=tips['tip_percentage'].mean(), color='red', linestyle='--', label=f'Среднее: {tips["tip_percentage"].mean():.1f}%')
plt.title('Процент чаевых от суммы счета')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip Percentage (%)')
plt.legend()

plt.tight_layout()
plt.show()

# 2. ВЛИЯНИЕ ВРЕМЕНИ ДНЯ
print(f"\n🕒 ВЛИЯНИЕ ВРЕМЕНИ ДНЯ:")
time_stats = tips.groupby('time', observed=True).agg({
    'total_bill': ['mean', 'std', 'count'],
    'tip': ['mean', 'std'],
    'tip_percentage': 'mean'
}).round(2)

print(time_stats)

# Визуализация времени дня
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(data=tips, x='time', y='total_bill', hue='time', legend=False)
plt.title('Распределение счетов по времени')
plt.xlabel('Время дня')
plt.ylabel('Total Bill ($)')

plt.subplot(1, 2, 2)
sns.boxplot(data=tips, x='time', y='tip', hue='time', legend=False)
plt.title('Распределение чаевых по времени')
plt.xlabel('Время дня')
plt.ylabel('Tip ($)')

plt.tight_layout()
plt.show()

# 3. РАСПРЕДЕЛЕНИЕ ПО ДНЯМ НЕДЕЛИ
print(f"\n📅 РАСПРЕДЕЛЕНИЕ ПО ДНЯМ НЕДЕЛИ:")
day_stats = tips.groupby('day', observed=True).agg({
    'total_bill': ['mean', 'count'],
    'tip': 'mean'
}).round(2)

print(day_stats)

# Сортировка по убыванию среднего счета
avg_by_day_sorted = tips.groupby('day', observed=True)['total_bill'].mean().sort_values(ascending=False)
print(f"\nСредние счета по дням (от большего к меньшему):")
for day, avg in avg_by_day_sorted.items():
    print(f"  {day}: ${avg:.2f}")

# Визуализация по дням
plt.figure(figsize=(10, 6))
sns.barplot(data=tips, x='day', y='total_bill', hue='day', estimator='mean', 
            errorbar=None, legend=False, order=['Sun', 'Sat', 'Thur', 'Fri'])
plt.title('Средние счета по дням недели')
plt.xlabel('День недели')
plt.ylabel('Средний Total Bill ($)')

# Добавляем значения на столбцы
ax = plt.gca()
for i, p in enumerate(ax.patches):
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 0.3,
            f'${height:.2f}', ha='center', va='bottom', fontweight='bold')

plt.show()

# 👥 ДЕМОГРАФИЧЕСКИЕ ПАТТЕРНЫ
print("\n4. 👥 ДЕМОГРАФИЧЕСКИЕ ПАТТЕРНЫ")
print("-" * 40)

# По полу (Sex)
print(f"\n🚻 АНАЛИЗ ПО ПОЛУ:")
sex_stats = tips.groupby('sex', observed=True).agg({
    'total_bill': ['mean', 'count'],
    'tip': 'mean',
    'tip_percentage': 'mean'
}).round(2)

print(sex_stats)

# Влияние курения (Smoker)
print(f"\n🚬 ВЛИЯНИЕ КУРЕНИЯ:")
smoker_counts = tips['smoker'].value_counts()
smoker_pct = tips['smoker'].value_counts(normalize=True) * 100

print(f"  • Курящие: {smoker_counts['Yes']} человек ({smoker_pct['Yes']:.1f}%)")
print(f"  • Некурящие: {smoker_counts['No']} человек ({smoker_pct['No']:.1f}%)")

smoker_analysis = tips.groupby('smoker', observed=True).agg({
    'total_bill': 'mean',
    'tip': 'mean',
    'tip_percentage': 'mean'
}).round(2)

print(smoker_analysis)

# Размер компании (Size)
print(f"\n👥 РАЗМЕР КОМПАНИИ:")
size_stats = tips['size'].value_counts().sort_index()
print("Распределение по размеру компании:")
for size, count in size_stats.items():
    pct = (count / len(tips)) * 100
    print(f"  • {size} человека: {count} посещений ({pct:.1f}%)")

# Визуализация демографических паттернов
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Пол и чаевые
sns.boxplot(data=tips, x='sex', y='tip', ax=axes[0, 0], hue='sex', legend=False)
axes[0, 0].set_title('Распределение чаевых по полу')
axes[0, 0].set_xlabel('Пол')
axes[0, 0].set_ylabel('Tip ($)')

# Курение и счета
sns.boxplot(data=tips, x='smoker', y='total_bill', ax=axes[0, 1], hue='smoker', legend=False)
axes[0, 1].set_title('Распределение счетов по статусу курения')
axes[0, 1].set_xlabel('Smoker')
axes[0, 1].set_ylabel('Total Bill ($)')

# Размер компании
size_dist = tips['size'].value_counts().sort_index()
axes[1, 0].bar(size_dist.index, size_dist.values, color='lightgreen', alpha=0.7, edgecolor='black')
axes[1, 0].set_title('Распределение по размеру компании')
axes[1, 0].set_xlabel('Размер компании')
axes[1, 0].set_ylabel('Количество посещений')

# Процент чаевых по полу и курению
sns.boxplot(data=tips, x='sex', y='tip_percentage', hue='smoker', ax=axes[1, 1])
axes[1, 1].set_title('Процент чаевых по полу и статусу курения')
axes[1, 1].set_xlabel('Пол')
axes[1, 1].set_ylabel('Tip Percentage (%)')
axes[1, 1].legend(title='Smoker')

plt.tight_layout()
plt.show()

# 📊 ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ
print("\n5. 📊 ДОПОЛНИТЕЛЬНЫЙ АНАЛИЗ")
print("-" * 40)

# Анализ выбросов
print(f"\n📊 АНАЛИЗ ВЫБРОСОВ:")
Q1 = tips['total_bill'].quantile(0.25)
Q3 = tips['total_bill'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = tips[(tips['total_bill'] < lower_bound) | (tips['total_bill'] > upper_bound)]
print(f"  • Выбросы в total_bill: {len(outliers)} записей ({len(outliers)/len(tips)*100:.1f}%)")

# Самые щедрые чаевые
top_tips = tips.nlargest(5, 'tip_percentage')[['total_bill', 'tip', 'tip_percentage', 'sex', 'day', 'time']]
print(f"\n🏆 САМЫЕ ЩЕДРЫЕ ЧАЕВЫЕ (топ-5 по проценту):")
print(top_tips.round(2))

# Статистические тесты
print(f"\n📊 СТАТИСТИЧЕСКИЕ ТЕСТЫ:")

# T-тест для чаевых по полу
male_tips = tips[tips['sex'] == 'Male']['tip']
female_tips = tips[tips['sex'] == 'Female']['tip']
t_stat, p_val = stats.ttest_ind(male_tips, female_tips)
print(f"  • T-тест чаевых по полу: p-value = {p_val:.4f} {'(значимо)' if p_val < 0.05 else '(не значимо)'}")

# T-тест для счетов по времени
lunch_bills = tips[tips['time'] == 'Lunch']['total_bill']
dinner_bills = tips[tips['time'] == 'Dinner']['total_bill']
t_stat_time, p_val_time = stats.ttest_ind(lunch_bills, dinner_bills)
print(f"  • T-тест счетов по времени: p-value = {p_val_time:.4f} {'(значимо)' if p_val_time < 0.05 else '(не значимо)'}")

# 📈 ИТОГОВЫЕ ВЫВОДЫ
print("\n" + "=" * 60)
print("🎯 КЛЮЧЕВЫЕ ВЫВОДЫ")
print("=" * 60)

print("""
📊 ОБЩАЯ ХАРАКТЕРИСТИКА:
• Данные качественные, без пропусков
• Высокая вариативность в суммах счетов и чаевых

💰 ФИНАНСОВЫЕ ПОКАЗАТЕЛИ:
• Средний процент чаевых: 15-16% - соответствует стандартной практике
• Высокий разброс значений указывает на разнородную клиентскую базу

📈 ВЗАИМОСВЯЗИ:
• Сильная корреляция счет-чаевые (0.67) - логичное поведение клиентов
• Выходные дни приносят наибольшую выручку
• Ужин - более доходное время, чем обед

👥 ДЕМОГРАФИЯ:
• Мужчины в среднем оставляют больше чаевых
• Статус курения не оказывает значительного влияния на финансовые показатели
• Наиболее популярный размер компании - 2 человека

🎯 РЕКОМЕНДАЦИИ ДЛЯ БИЗНЕСА:
1. Усилить маркетинг в выходные дни
2. Разработать программы лояльности для постоянных клиентов
3. Оптимизировать staffing под пиковые нагрузки (ужин, выходные)
""")

# Финальная визуализация - сводная информация
plt.figure(figsize=(14, 10))

# 1. Распределение основных показателей
plt.subplot(2, 2, 1)
sns.histplot(tips['total_bill'], kde=True, color='skyblue')
plt.title('Распределение Total Bill')
plt.xlabel('Total Bill ($)')

plt.subplot(2, 2, 2)
sns.histplot(tips['tip'], kde=True, color='lightcoral')
plt.title('Распределение Tip')
plt.xlabel('Tip ($)')

# 2. Процент чаевых
plt.subplot(2, 2, 3)
sns.histplot(tips['tip_percentage'], kde=True, color='lightgreen')
plt.axvline(tips['tip_percentage'].mean(), color='red', linestyle='--', 
            label=f'Среднее: {tips["tip_percentage"].mean():.1f}%')
plt.title('Распределение процента чаевых')
plt.xlabel('Tip Percentage (%)')
plt.legend()

# 3. Сводная тепловая карта корреляций
plt.subplot(2, 2, 4)
numeric_cols = ['total_bill', 'tip', 'size', 'tip_percentage']
corr_matrix = tips[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, fmt='.2f', cbar_kws={'shrink': 0.8})
plt.title('Матрица корреляций')

plt.tight_layout()
plt.show()

print("✅ Анализ завершен! Слава роботам!")
```
