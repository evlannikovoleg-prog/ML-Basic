```python
import sympy

# Определяем переменные питона в качестве математических объектов SymPy
x, y, z, v = sympy.symbols('x y z v')

# Создаём матрицы:
A = sympy.Matrix([[x, 2, 3],
                  [-1, y, 4]])

B = sympy.Matrix([[1, 2, -5],
                  [2, -6, z]])

C = sympy.Matrix([[8, v, -1],
                  [1, 6, 4]])

O = sympy.Matrix([[0, 0, 0],
                  [0, 0, 0]])

# Решаем уравнение: 3*A + 2*B = C

print("3A: ", 3*A)
print("2B: ", 2*B)

equation_matrix = 3*A + 2*B - C 
print("equation_matrix: ", equation_matrix)

# Собираем уравнения из каждого элемента
equations = []
for i in range(equation_matrix.rows):
    for j in range(equation_matrix.cols):
        equations.append(equation_matrix[i, j])
print(equations)

# Решаем систему уравнений
solution = sympy.solve(equations, (x, y, z, v), dict=True)
print(solution)
```

```python
%%latex
$$
Ответ:
$$$$
x=2,\ y=6,\ z=-4,\ v=10.
$$
```

```python
# Исследовательское задание: малоранговая аппроксимация матрицы. Сгенерируйте случайную квадратную матрицу A ∈ R
# n×n, n ≥ 100. С помощью функции numpy.linalg.svd выполните сингулярное разложение этой матрицы, и получите три матрицы: U, S, V
# ⊤. Выполняйте аппроксимацию матрицы A с
# рангом r, меняя его значение, например, от 2 до n...
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
n = 100  # размер квадратной матрицы (n >= 100)

# 1. Генерация случайной квадратной матрицы
np.random.seed(42)  # для воспроизводимости результатов
A = np.random.randn(n, n)

# 2. Сингулярное разложение (SVD)
U, S_diag, Vt = np.linalg.svd(A, full_matrices=False)

# Преобразуем S в диагональную матрицу для удобства
S = np.diag(S_diag)

# 3. Подготовка массивов для хранения ошибок
ranks = np.arange(1, n + 1)  # ранги от 1 до n
errors = np.zeros(n)

# 4. Вычисление ошибок для каждого ранга r
for r in ranks:
    # Малоранговая аппроксимация
    A_approx = U[:, :r] @ S[:r, :r] @ Vt[:r, :]

    # Ошибка аппроксимации (норма Фробениуса)
    error = np.linalg.norm(A - A_approx, 'fro')
    errors[r - 1] = error

# 5. Построение графика
plt.figure(figsize=(10, 6))
plt.plot(ranks, errors, 'b-', linewidth=2)
plt.xlabel('Ранг аппроксимации (r)', fontsize=12)
plt.ylabel('Ошибка аппроксимации E(r)', fontsize=12)
plt.title('Зависимость ошибки малоранговой аппроксимации от ранга', fontsize=14)
plt.grid(True, alpha=0.3)

# Добавим вертикальную линию для визуализации
# Например, покажем ранг, при котором достигается 90% точности
total_norm = np.linalg.norm(A, 'fro')
target_error = 0.1 * total_norm  # 10% от общей нормы
idx = np.where(errors <= target_error)[0]
if len(idx) > 0:
    r_90 = idx[0] + 1
    plt.axvline(x=r_90, color='r', linestyle='--', alpha=0.7,
                label=f'90% точности при r={r_90}')

plt.legend()
plt.tight_layout()
plt.show()

# 6. Дополнительная информация
print(f"Размер матрицы: {n}×{n}")
print(f"Норма Фробениуса исходной матрицы: {total_norm:.4f}")
print(f"Ошибка при r=1: {errors[0]:.4f}")
print(f"Ошибка при r=10: {errors[9]:.4f}")
print(f"Ошибка при r=50: {errors[49]:.4f}")
print(f"Ошибка при полном ранге (r={n}): {errors[-1]:.4e}")

# 7. Построение графика в логарифмическом масштабе (дополнительно)
plt.figure(figsize=(10, 6))
plt.semilogy(ranks, errors, 'g-', linewidth=2)
plt.xlabel('Ранг аппроксимации (r)', fontsize=12)
plt.ylabel('log(Ошибка аппроксимации E(r))', fontsize=12)
plt.title('Зависимость ошибки аппроксимации от ранга (логарифмическая шкала)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
